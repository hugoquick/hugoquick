```python
import numpy as np
from sympy import*
import matplotlib.pyplot as plt



n=9 # numero de pontos

y = np.linspace(-2,2,n)
t = np.linspace(-2,2,n)
Y,T = np.meshgrid(y,t)
# Y,T posições
#dy,dt direção da seta
#plt.scatter(Y,T)

f = y+T

angulo = np.arctan(f)

dy = np.cos(angulo)
dt = np.sin(angulo)

plt.quiver(Y,T,dy,dt)
```
